```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted", font_scale=1.1)

df = pd.read_csv("../data/DT_Store_Suplements_final.csv")
df["Receita"] = df["Preco"] * df["Quantidade"]
```

## ðŸ”¹ Receita Total

```python
receita_total = df["Receita"].sum()
print(f"Receita total: R$ {receita_total:,.2f}")
```

ðŸ‘‰ **Receita Total:** O faturamento total no perÃ­odo analisado foi de **R$ XXX.XXX,XX**.

## ðŸ”¹ Top 5 Produtos mais Vendidos (Quantidade)

```python
top_produtos_qtd = df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_produtos_qtd.values, y=top_produtos_qtd.index, palette="Blues_r")
plt.title("Top 5 Produtos mais Vendidos (Quantidade)")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")
plt.show()
```

ðŸ‘‰ O produto mais vendido em quantidade foi **XYZ**, seguido por **ABC**.

## ðŸ”¹ Top 5 Produtos por Receita

```python
top_produtos_receita = df.groupby("Produto")["Receita"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_produtos_receita.values, y=top_produtos_receita.index, palette="Greens_r")
plt.title("Top 5 Produtos por Receita")
plt.xlabel("Receita (R$)")
plt.ylabel("Produto")
plt.show()
```

ðŸ‘‰ Em faturamento, o destaque foi o **XYZ**, responsÃ¡vel por grande parte da receita.

## ðŸ”¹ Marcas que mais faturam

```python
top_marcas_receita = df.groupby("Marca")["Receita"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_marcas_receita.values, y=top_marcas_receita.index, palette="Oranges_r")
plt.title("Top Marcas por Receita")
plt.xlabel("Receita (R$)")
plt.ylabel("Marca")
plt.show()
```

ðŸ‘‰ A marca **AAA** foi a que mais faturou, mostrando forte presenÃ§a no mercado.

## ðŸ”¹ DistribuiÃ§Ã£o de Formas de Pagamento

```python
plt.figure(figsize=(6,6))
df["Pagamento"].value_counts().plot.pie(
    autopct="%.1f%%", startangle=90, colors=sns.color_palette("pastel")
)
plt.title("DistribuiÃ§Ã£o de Formas de Pagamento")
plt.ylabel("")
plt.show()
```

ðŸ‘‰ Observa-se que **CartÃ£o** Ã© a forma de pagamento mais usada, seguido de **PIX**.

## ðŸ”¹ ComparaÃ§Ã£o entre Lojas

```python
lojas_receita = df.groupby("Loja")["Receita"].sum()

plt.figure(figsize=(7,5))
sns.barplot(x=lojas_receita.index, y=lojas_receita.values, palette="Purples_r")
plt.title("Receita por Loja")
plt.xlabel("Loja")
plt.ylabel("Receita (R$)")
plt.show()
```

ðŸ‘‰ A **Loja X** apresentou o maior faturamento, enquanto a **Loja Y** teve desempenho inferior.
